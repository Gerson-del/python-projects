```python
import random
import time

def Even_Odd(n):
  if n % 2 == 0:
    return "Your number is even"
  return "your number is odd"

def guess_number():
    Max_attempts = 10

    print("Welcome to the Number Guessing Game!")
    print("Choose the difficulty:")
    print("1. Easy (Numbers between 0 and 9)")
    print("2. Medium (Numbers between 0 and 15)")
    print("3. Hard (Numbers between 0 and 25)")

    #settings:

    while True:
        print("Settings")
        difficulty = int(input("Choose the difficulty (1 easy, 2 medium, 3 hard): "))
        time_limit = int(input("Choose the time limit: 20 seconds - 50 seconds: "))
        if difficulty == 1:
            d = 9
        elif difficulty == 2:
            d = 15
        elif difficulty == 3:
            d = 25

        if 1 <= difficulty <= 3 and 20 <= time_limit <= 50:
            print(f"The number is between 0 and {d}")
            break
        else:
            print("Please enter a valid number")



    while True:
        start_time = time.time()  # Reset start_time for each game
        n = random.randint(0, d)
        attempts = 0

        #This will give you a hint if you've failed five times







        while True:
            if attempts == 10:
              print("You've ran out of attempts")
              break
            elif attempts % 5 == 0 and attempts != 0:
              hint = input("it appears you're having trouble finding the answer, do you want a hint? (yes/ no) : ").lower()
              if hint == "yes":
                print(Even_Odd(n))

            elapsed_time = time.time() - start_time
            if elapsed_time > time_limit:
                print("Time's up, better luck next time ")
                break

            guess = input(f"Time remaining {int(time_limit - elapsed_time)}, Make a guess: (Type exit to quit): ")
            if guess.lower() == "exit":
                break
            try:
                guess = int(guess)
            except ValueError:
                print("Invalid input. Please enter a number.")
                continue

            if guess == n:
                print("Correct! You guessed the number.")
                print("-----------------")
                print(f"Score: {(Max_attempts - attempts) * 10}%")
                print("-----------------")
                break
            elif guess > n:
                print("Too high :c, try again")
            elif n > guess:
                print("Too low, try again")

            attempts += 1

        play_again = input("Do you want to play again? (yes/no): ")
        if play_again.lower() != "yes":
            break

if __name__ == "__main__":
    guess_number()
```
